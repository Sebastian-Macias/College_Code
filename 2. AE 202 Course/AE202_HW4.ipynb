```python
#Problem 2

import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
import math as m

#a
e = 0
re = 6371*1000 #km to m
G = 6.6743*10**-11 #m^3/kg*s^2
Me = 5.972*10**24 #kg

def g(h):
    r = h+re
    g = (G*Me)/(r**2)
    return g

h = np.linspace(200*1000, 40000*1000, 1000)
plt.plot(h, g(h))
plt.ylabel("Gravitational accel (m/s^2)")
plt.xlabel('Altitude(m)')
plt.title('Gravitational accel (m/s^2) vs Altitude (m)')
plt.show()

#b
e = 0
re = 6371 #km
G = 6.6743*10**-20 #km^3/kg*s^2
Me = 5.972*10**24 #kg

def v(h):
    r = h+re
    v = ((G*Me)/(r))**.5
    return v

h = np.linspace(200, 40000, 1000)
plt.plot(h, v(h))
plt.ylabel("Orbital Velocity (km/s)")
plt.xlabel('Altitude (km)')
plt.title('Orbital Velocity (km/s) vs Altitude (km))')
plt.show()

#c
e = 0
re = 6371 #km
mue = 3.986*10**5 #km^3/s^2

def Tp(h):
    r = h+re
    Tp = (2*m.pi)/(m.sqrt(mue*(3600**2)))*(r**(3/2))
    return Tp

h = np.linspace(200, 40000, 1000)
plt.plot(h, Tp(h))
plt.ylabel("Orbital Period (hr)")
plt.xlabel('Altitude (km)')
plt.title('Orbital Period (hr) vs Altitude (km))')
plt.show()

#show value at set time
Tsidday =(23*3600)+(56*60)+4
rnew = (Tsidday/(2*m.pi/m.sqrt(mue)))**(1/(3/2))
print(rnew)
print(rnew-re) #to show within linspace of 200 to 40,000
```

```python
#problem 3
#a
mus = 1.327*10**11 #km^3/s^2
rp = np.array([46001009,107476170,147098291,206655215,740679835,1349823615,2734998229,4459753056])
ra = np.array([69817445,108942780,152098233,249232432,816001807,1503509229,3006318143,4537039826])

def T(a):
    T = 2*m.pi/m.sqrt(mus)*(a)**(3/2)
    return T/(3600*24*365) #units years

mean = (rp+ra)/2
period = T(mean)

plt.loglog(period, mean, 'o')
plt.ylabel("Semimajor axis (km)")
plt.xlabel('Period (years)')
plt.title('Semimajor axis (km) vs Period (years))')
plt.show()

#b
names = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune"]
Ratio = (mean**3)/(period**2)

plt.plot(names,Ratio, 'o')
plt.xlabel("Planet Names")
plt.ylabel('Keplers 3rd Law Ratio')
plt.title('Planet Names vs Keplers 3rd Law Ratio')
plt.show()

#c
plutorp = 4436756954 #km
plutora = 7376124302 #km

curvefit = np.polyfit(np.log(period), np.log(mean), deg = 1)

meanpluto = (plutorp+plutora)/2

def plutoperiod(meanpluto):
    T = np.exp((np.log(meanpluto)-curvefit[1])/(curvefit[0]))
    return T

print(plutoperiod(meanpluto))

print('curve fit period', plutoperiod((plutorp+plutora)/2))
print('calculated period', T((plutorp+plutora)/2))
```
