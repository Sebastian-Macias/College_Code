```python
#Question 6.a
import numpy as np
import matplotlib.pyplot as plt
import math as m
MachDen = np.linspace(0, 1, 100)

gamma = 1.4

def rhoflow(M):
    rho = (1 + ((gamma-1)/2) * M**2)**(1/(gamma-1))
    return rho
M = np.linspace(0, 1, 100)
plt.plot(M, rhoflow(M))
plt.ylabel("Density Ratio (rho0/rho)")
plt.xlabel('Mach Number (M)')
plt.title('Mach Number (M) vs. Density Ratio (rho0/rho)')
plt.show()

#Problem 6.b
print(rhoflow(0.3))


```
