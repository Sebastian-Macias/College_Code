```python
#Problem 5

import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
import math as m

cm0 = 0.08
sref = 53.3 #ft^2
st = 14.2 #ft^2
c = 5 #ft
lt = 36 #ft
it = 0
e0 = 1.8
deda = 0.35
aw = 0.09 #/deg
cmcgwb = -0.032

#5a
#assumption 1: trim absolute angle of attack is absolute angle of attack with trim conditions
alphaw = 4 #/deg
VH = (lt*st)/(sref*c)
E = e0 + (deda*alphaw)
at = alphaw - E - it
cmacwb = cm0 - (VH*at*(it+e0))
H = (cmcgwb-cmacwb)/(aw*alphaw)
#H = h-hacwb
print(H)


#5B
#Assumption 2: Alphat = at bcuz there is no other way to solve
#alphaw = abs angle of attack 
#make alphaw a list of 0 to 10, cmcgwb is constant which matters for H = h-hacwb
#cmcg = cmacwb + (aw*alphaw1*(H-(VH*(at/aw)*(1-deda)))) + (VH*at*(it+e0))

def cmcg(alphaw1):
    cmcg = cmacwb + (aw*alphaw1*(H-(VH*(at/aw)*(1-deda)))) + (VH*at*(it+e0))
    return cmcg

alphaw1 = np.linspace(0, 10, 100)
plt.plot(alphaw1, cmcg(alphaw1))
plt.xlabel("Angle of attack (in degrees)")
plt.ylabel('Moment coef with respect to c.g (Cm,c.g)')
plt.title('Moment coef with respect to c.g (Cm,c.g) vs Angle of attack (in degrees)')
plt.show()
```
