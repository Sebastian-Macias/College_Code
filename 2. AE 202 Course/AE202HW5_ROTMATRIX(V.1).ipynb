```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
import math as m
import numpy.linalg as la
import scipy.linalg as sla

#Problem 1a

#OMEGA = Captital omega 
#w = lowercase omega
#i = i

#Orbit 1
a = 7371.0 #km
e = 0.1 
i = 28.5 * (m.pi/180) #deg to rad
w = 191 * (m.pi/180) #deg to rad
OMEGA = 273 * (m.pi/180) #deg to rad
f = 10 * (m.pi/180) #deg to rad
r = np.array([-2413.9, 6083.8, -1136.0]) #km
v = np.array([-6.5461, -3.1365, -3.6385]) #km/s

print("i =", i* 180/m.pi)
print("OMEGA =", OMEGA* 180/m.pi)
print("w =", w* 180/m.pi)
print("f =", f* 180/m.pi)
print("a =",a)
print("e =",e)
print(      )
print("r =", r)
print("v =",v)
print(       )
#####Classical Orbital Elements (COE) to r,v#####

#when testing use OMEGA = 0, w = 0, i = 0 to get identity matrix
RotMatrix = np.array([[m.cos(OMEGA)*m.cos(w) - m.sin(OMEGA)*m.sin(w)*m.cos(i) , -m.cos(OMEGA)*m.sin(w) - m.sin(OMEGA)*m.cos(w)*m.cos(i), m.sin(OMEGA)*m.sin(i)],
    [m.sin(OMEGA)*m.cos(w) + m.cos(OMEGA)*m.sin(w)*m.cos(i), -m.sin(OMEGA)*m.sin(w) + m.cos(OMEGA)*m.cos(w)*m.cos(i), -m.cos(OMEGA)*m.sin(i)],
    [m.sin(w)*m.sin(i), m.cos(w)*m.sin(i), m.cos(i)]])

print("Rotation Matrix =", RotMatrix)

muE = 3.986*10**5 #km^3/s^2

def rpqw(a,e,f):
    P = a*(1-(e**2))
    r1 = P/(1+e*m.cos(f))
    rpqw = np.array([r1*m.cos(f), r1*m.sin(f),0])
    return rpqw

def vpqw(a,e,f):
    P = a*(1-(e**2))
    vpqw = np.array([m.sqrt(muE/P)*-m.sin(f), m.sqrt(muE/P)*(e+m.cos(f)), 0])
    return vpqw

print(    )
print("rpqw =", rpqw(7371.0,0.1,10 * (m.pi/180)))
print("vpqw =", vpqw(7371.0,0.1,10 * (m.pi/180)))

rvec = RotMatrix@rpqw(7371.0,0.1,10 * (m.pi/180))
vvec = RotMatrix@vpqw(7371.0,0.1,10 * (m.pi/180))

print(      )
print("rvec =", rvec)
print("vvec =", vvec)

#####r,v to COE#####
print(      )

hvec = np.cross(rvec,vvec) #cross product func
print("angular vel vec=", hvec)

kvec = np.array([0,0,1])
nvec = np.cross(kvec,hvec) #n cross k func
nmag = np.linalg.norm(nvec)
print("nmag =", nmag)
print("Normal Vec =", nvec)

vmag = np.linalg.norm(vvec) #magnitude of vvec
rmag = np.linalg.norm(rvec) #magnitude of rvec

print("rmag =",rmag)
print("vmag =", vmag)

evec = 1/muE * ((vmag**2 - muE/rmag)*rvec - (rvec@vvec)*vvec)
emag = np.linalg.norm(evec)

print("eccentricity vec =", evec)
print("eccentricity mag =", emag)
print(      )

P = (np.linalg.norm(hvec)**2)/muE
a = P/(1-emag**2)

print("a =",a)
print(    )

## take h,n,e vectors and find OMEGA, w, i
h_k = np.linalg.norm(kvec@hvec)
hmag = np.linalg.norm(hvec)
i = m.acos(h_k/hmag)
print("angle of inclination (i) =",i * 180/m.pi)

I = [1,0,0] #i direction

if nvec[1]>0:
    OMEGA = m.acos(nvec[0]/nmag) 
else:
    OMEGA = 2*m.pi - m.acos(nvec[0]/nmag)

print("OMEGA angle =", OMEGA * 180/m.pi)

ndot = np.dot(nvec,evec)
emag = np.linalg.norm(evec)

if evec[2]>0:
    w = m.acos(ndot/(nmag*emag))
else: 
    w = 2*m.pi - m.acos(ndot/(nmag*emag))
print("argument of periapsis (w) =", w * 180/m.pi)

e_f = evec@rvec
if np.dot(rvec,vvec) > 0:
    f = m.acos(e_f/(emag*np.linalg.norm(rvec)))
else:
    f = 2*m.pi - m.acos(e_f/(emag*np.linalg.norm(rvec)))
print("True anomaly at epoch (f) =",f * 180/m.pi)
```
