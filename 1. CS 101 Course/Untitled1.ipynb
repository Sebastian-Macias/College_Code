```python
career_rbis = {}

# Open the file.  Call it batting_file.

batting_file = open('Batting.csv','r')

# Read the data from the file using a `for` loop.
for line in batting_file.readlines():
    line = line.strip()  # remove whitespace from the line
    values = line.split(",")  # split the line by commas ','
    if values[ 0 ] == 'playerID':
        continue  # skip the header line
    player_id = values[0]
    try:
        rbi = float(values[12])
    except:
        continue
    if player_id not in career_rbis.keys():
        career_rbis[player_id] = rbi
    else:
        career_rbis[player_id] += rbi

max_player = ''
max_rbis = 0
for key, value in career_rbis.items():
    if career_rbis[key] > max_rbis:
        max_rbis = career_rbis[key]
        max_player = key
        
print(max_player)
print(max_rbis)
```
