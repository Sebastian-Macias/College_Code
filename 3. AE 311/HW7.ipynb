```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
import math as m
import numpy.linalg as la
import scipy.linalg as sla

csv = open('naca24112-jf.csv')
lines = csv.readlines()

x = []
y = []

x_airfoil = []
x_camber = []
x_chord =  []
y_airfoil = []
y_camber = []
y_chord = []


for i in lines:
    i.split(',')
    try:
        x.append(float(i.split(',')[0].strip()))
        y.append(float(i.split(',')[1].strip()))
    except ValueError:
        continue
    
    x_airfoil = x[0:121]
    y_airfoil = y[0:121]
    x_camber = x[121:184]
    y_camber = y[121:184]
    x_chord = x[184:186]
    y_chord = y[184:186]
    
plt.plot(x_airfoil, y_airfoil, label='Airfoil')
plt.plot(x_camber, y_camber, label='Camber')
plt.plot(x_chord, y_chord, label='Chord')
plt.title('NACA 0012')
plt.xlabel('length')
plt.ylabel('y axis')
plt.ylim(-30,30)
plt.legend()
plt.show()
```
