```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
import math 
import numpy.linalg as la
import scipy.linalg as sla
import csv
import matplotlib 

#Problem 11 
#u quiver plot

from numpy import genfromtxt
u = genfromtxt('u.csv', delimiter=',', encoding='utf-8-sig')[:,:-1]
v = genfromtxt('v.csv', delimiter=',', encoding='utf-8-sig')
x = genfromtxt('x.csv', delimiter=',', encoding='utf-8-sig')
y = genfromtxt('y.csv', delimiter=',', encoding='utf-8-sig')

X, Y = np.meshgrid(x, y)

plt.contourf(X,Y,u)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('u')
cbar = plt.colorbar()
cbar.set_label('U colorbar')
plt.show() #filled contour plot

#v quiver plot

plt.contourf(X,Y,v)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('v')
cbar = plt.colorbar()
cbar.set_label('V colorbar')
plt.show() #filled contour plot

#combined u and v plot 

plt.quiver(X,Y,u,v)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('u and v')
plt.show()
```

```python
#problem 12
gradx = np.gradient(u,axis = 1)
grady = np.gradient(v,axis = 0)
div = gradx + grady

plt.contourf(X,Y,div)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Divergence')
cbar = plt.colorbar()
cbar.set_label('Divergence colorbar')
plt.show() #filled contour plot

print(np.sum(div))

```
