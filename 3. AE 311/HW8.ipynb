```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
import math as m
import numpy.linalg as la
import scipy.linalg as sla

a = 10* np.pi/180
V = 10 #m/s

theta = np.linspace(0,np.pi,200)

gamma = 2*a*V * (1+np.cos(theta))/np.sin(theta)


plt.plot(theta,gamma)
plt.xlabel('Theta (rad)')
plt.ylabel('Circulation Strength at Given Theta')
plt.title('Local Circulation Strength')
plt.show() 
```
