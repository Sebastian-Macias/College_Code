```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,20)
y = np.linspace(-15,15,20)
z = np.linspace(-10,10,20)
X,Y,Z = np.meshgrid(x,y,z)


i = 2 * X * Y
j = 4 * Y ** 2
k = 5 * Z

dx = x[1]-x[0]
dy = y[1]-y[0]
dz = z[1]-z[0]

gradx = np.gradient(i,dx,axis=1)
grady = np.gradient(j,dy,axis=0)
gradz = np.gradient(k,dz,axis=2)

div = gradx+grady+gradz
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import matplotlib

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


# Creating color map
my_cmap = plt.get_cmap('jet')
 
# Creating plot
list= []
for i in range(20):
    for j in range(20):
        for k in range(20):
            list.append(div[j][i][k])
            


sctt = ax.scatter3D(X, Y, Z,
                    alpha = 0.8,
                    c=list, 
                    cmap = my_cmap, 
                    marker ='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

#fig.colorbar(sctt,ax=ax)

import matplotlib.cm as cmx
cNorm = matplotlib.colors.Normalize(vmin=np.min(list), vmax=np.max(list))
scalarMap = cmx.ScalarMappable(norm=cNorm, cmap=my_cmap)
cbar = plt.colorbar(scalarMap,fraction=0.2, pad=0.1,shrink=1)
cbar.ax.set_ylabel('w$_{2}$')
cbar.ax.autoscale_view()

plt.show()

plt.contourf(X, Y, 2, 20, cmap='RdGy')
plt.colorbar();
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,20)
y = np.linspace(-15,15,20)
z = np.linspace(-10,10,20)
X,Y,Z = np.meshgrid(x,y,z)


i = 2 * X * Y
j = 4 * Y ** 2
k = 5 * Z

dx = x[1]-x[0]
dy = y[1]-y[0]
dz = z[1]-z[0]

gradx = np.gradient(i,dx,axis=1)
grady = np.gradient(j,dy,axis=0)
gradz = np.gradient(k,dz,axis=2)

divNum = gradx+grady+gradz
divNum = divNum[:,:,0]

X,Y = np.meshgrid(x,y)

plt.contourf(X,Y,divNum)
plt.colorbar()
plt.title('Numerical Soln to Divergence')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
