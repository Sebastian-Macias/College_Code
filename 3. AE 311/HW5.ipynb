```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
import math as m
import numpy.linalg as la
import scipy.linalg as sla
import csv
import matplotlib 

x = np.linspace(-10,10,200)
y = np.linspace(-10,10,200)

X, Y = np.meshgrid(x, y)

V = -2*Y - 2*X 
psi = X**2 - Y**2
phi = -2*X*Y

plt.contour(X,Y,psi)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Streamline Function')
plt.colorbar().set_label(('Stream Function Magnitude'))
plt.show()

plt.contour(X,Y,phi)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Potential Function')
plt.colorbar().set_label(('Potential Function Magnitude'))
plt.show()


plt.contour(X,Y,psi,colors = 'g')
plt.contour(X,Y,phi,colors = 'r')
plt.xlabel('x axis')
plt.ylabel('y axis')
stream = matplotlib.patches.Patch(color= 'green', label = 'Streamlines') 
potential = matplotlib.patches.Patch(color= 'red', label = 'Potential lines') 
plt.legend(handles = [stream, potential])
plt.title('Streamline and Potential Function')
plt.show()
```
